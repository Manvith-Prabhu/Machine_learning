# Filtering text

```python
def remove_non_ascii(text):
    #Remove non-ASCII characters from list of tokenized words
    return text.encode('ascii','ignore').decode()

def remove_brackets_num(text):
    return re.sub("","",text)

def to_lowercase(text):
    return text.lower()

def remove_whitespace(text):
    return text.strip()

def remove_punctuation(text):
    punctuation= '''!()[]{};:'"\<>/?$%^&.*_`~='''
    for punc in punctuation:
        text=text.replace(punc,"")
    return text

def remove_emails(text):
    return re.sub(r'[A-Za-z0-9]*@[A-Za-z]*\.?[A-Za-z0-9]*', "", text)

def text2words(text):
    return word_tokenize(text)

def remove_stopwords(words,stop_words):
    return [word for word in words if word not in stop_words]
```

```python
def normalize_text(text):
    text = remove_non_ascii(text)
    text= remove_brackets_num(text)
    text = to_lowercase(text)
    text= remove_whitespace(text)
    text = remove_punctuation(text)
    text= remove_emails(text)
    words = text2words(text)
    return ' '.join(words)
```

```python
clean_data = normalize_text(text)
```

# Keywords Extraction

```python
def topicRank(text):
    extractor=pke.unsupervised.TopicRank()                # Using the TopicRank keyphrase extraction model
    extractor.load_document(input=text, language='en')    # loading the text
    extractor.candidate_selection()                       # Sequences of noun and adjectives are created
    extractor.candidate_weighting()                       # Ranking the keyphrases
    keyphrases= extractor.get_n_best(n=10)                # 10 best keyphrases are selected
    return [a_tuple[0] for a_tuple in keyphrases]
```

```python
def get_question(sentence,answer,mdl,tknizer):
    text = "context: {} answer: {}".format(sentence,answer)
    max_len = 512
    encoding = tknizer.encode_plus(text,max_length=max_len, pad_to_max_length=False,truncation=True, return_tensors="pt")
    input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
    outs = mdl.generate(input_ids=input_ids,
                                attention_mask=attention_mask,
                                early_stopping=True,
                                num_beams=5,
                                num_return_sequences=1,
                                no_repeat_ngram_size=2,
                                max_length=300)
    dec = [tknizer.decode(ids,skip_special_tokens=True) for ids in outs]
    Question = dec[0].replace("question:","")
    Question= Question.strip()
    return Question
```

```python
from transformers import RobertaTokenizer
from transformers import RobertaForMultipleChoice
import torch
from torch.distributions import Categorical
import itertools as it
import nlp2go

tokenizer = RobertaTokenizer.from_pretrained("LIAMF-USP/roberta-large-finetuned-race")
model = RobertaForMultipleChoice.from_pretrained("LIAMF-USP/roberta-large-finetuned-race")
model.eval()
model.to("cuda")

dg_model = nlp2go.Model('./BDG.pt')
dg_model_pm = nlp2go.Model('./BDG_PM.pt')
dg_model_both = nlp2go.Model('./BDG_ANPM.pt')
```

```python
context = text
all_options=[]
for i in range(len(ques)):
  d_input = context + '' + ques[i] + '' + ans[i]
  choices = dg_model.predict(d_input, decodenum=3)['result']
  choices_pm = dg_model_pm.predict(d_input, decodenum=3)['result']
  choices_both = dg_model_both.predict(d_input, decodenum=3)['result']
  all_options.append(choices + choices_pm + choices_both)
```

```python
import os
import itertools as it
import torch
from torch.distributions.categorical import Categorical
from nlgeval import NLGEval

nlgeval = NLGEval(
    metrics_to_omit=['METEOR', 'EmbeddingAverageCosineSimilairty', 'SkipThoughtCS', 'VectorExtremaCosineSimilarity',
                     'GreedyMatchingScore', 'CIDEr'])
os.environ['CUDA_VISIBLE_DEVICES'] = '1'

def selection(context, question, answer, all_options):
    max_combin = [0, []]
    for combin in set(it.combinations(all_options, 3)):
        options = list(combin) + [answer]
        keep = True
        for i in set(it.combinations(options, 2)):
            a = "".join([char if char.isalpha() or char == " " else " " + char + " " for char in str(i[0].tolist())])
            b = "".join([char if char.isalpha() or char == " " else " " + char + " " for char in str(i[1].tolist())])
            metrics_dict = nlgeval.compute_individual_metrics([a], b)

            if metrics_dict['Bleu_1'] > 0.5:
                keep = False
                break
        if keep:
            prompt = context + tokenizer.sep_token + question
            encoding_input = []
            for choice in options:
                encoding_input.append([prompt, choice])
            encoding_input.append([prompt, answer])
            labels = torch.tensor(len(options) - 1).unsqueeze(0)
            encoding = tokenizer(encoding_input, return_tensors='pt', padding=True, truncation='only_first')
            outputs = model(**{k: v.to('cuda:1') for k, v in encoding.items()},
                            labels=labels.to('cuda:1'))  # batch size is 1
            entropy = Categorical(probs=torch.softmax(outputs.logits, -1)).entropy().tolist()[0]
            if entropy >= max_combin[0]:
                max_combin = [entropy, options]
    return max_combin[1][:-1]
```
